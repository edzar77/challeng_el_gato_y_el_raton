    El algoritmo minimax es un algoritmo de toma de decisiones para juegos de dos jugadores con turnos alternos (como ajedrez, tres en raya o damas) que busca encontrar la mejor jugada posible. Funciona asumiendo que el oponente siempre realizará su mejor movimiento para perjudicarlo, por lo que el jugador "Max" intenta maximizar su puntuación, mientras que el jugador "Min" intenta minimizar la puntuación del oponente. Para hacer esto, el algoritmo explora recursivamente un árbol de juego, evaluando todas las posibles jugadas futuras y asignando valores a las posiciones finales para elegir el camino óptimo
    -algoritmos de decision
        el gato va a maximizar el resultado de las decisiones 
        el raton va a minimizar el resutado de las desiciones
        si, el queso es comido por el raton, el raton gana.
        si, el raton es comido por el gato, el gato gana.
